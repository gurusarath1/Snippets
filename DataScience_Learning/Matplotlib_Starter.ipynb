```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,20)
y1 = x*x + 3
y2 = np.sin(x)*100
print(x)
print(y1)
print(y2)
```

## Matlab Style

```python
plt.figure(figsize=(10,10))
plt.plot(x, y1, c='r', marker='o', label='Quadratic')
plt.plot(x, y2, c='g', label='Sine')
plt.title("This is the title of the graph", fontsize=20)
plt.xlabel("X Label", fontsize=20)
plt.ylabel("Y Label", fontsize=20)
plt.grid()
plt.legend(loc='upper right', fontsize= 20) # You need to set label inorder for the legend to appear
plt.annotate("This is location 0,0 ", (0,0))
plt.xlim(left=0, right=15)
plt.ylim(top=150, bottom=-100)
plt.xticks(rotation=90)
plt.yticks([0,25,50,75,100], ['Zero','mid','upper mid', 'lower high', 'high'])
```

## Object Oriented Style

```python
fig = plt.figure(figsize=(20,10))


ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

plt.suptitle('TITLE FOR BOTH', fontsize=20)

ax1.set_title("Quadratic")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.plot(x, y1, c='b')
ax1.grid()

ax2.set_title("Sine")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.plot(x, y2, c='g')
ax2.grid(axis='y')  #axis = {both, x, y}
```
