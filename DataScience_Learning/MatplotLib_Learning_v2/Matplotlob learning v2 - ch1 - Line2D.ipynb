```python
import numpy as np
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt

%matplotlib inline
```

# Line2D

```python
#Generate some random data
np.random.seed(10)

x = np.linspace(1, 40, 10)

y1 = np.sin(x);
y2 = np.cos(x) + np.sqrt(x*4)
```

```python
fig, ax = plt.subplots()

#create two lines
line1 = Line2D(x, y1)
line2 = Line2D(x, y2)

# Add the line to the axis
ax.add_line(line1)
ax.add_line(line2)
```

```python
ax.autoscale()
fig
```

```python
# Get the list of all the line
ax.lines
```

```python
#referencing an individual line
ax.lines[0].set_color('red')
ax.lines[1].set_color('b')

fig
```

```python
#create handlers for lines
l1 = ax.lines[0]
l2 = ax.lines[1]
```

### Line2D : Styling Line2D

```python
l1.set_alpha(0.2)
l1.set_linestyle('dashed')

fig
```

```python
l2.set_marker('P')
l2.set_markersize(10)
l2.set_markerfacecolor('yellow')
fig
```

```python
l2.set_markevery(2)
fig
```

```python
l2.set_sketch_params(0.5) #line will look a bit sketchy
fig
```

```python
l1.set_clip_on(False) # prevents line clipping when it goes out of limits of axis
ax.set_xlim(0,25) # change the axis range
fig
```

```python
fig.savefig('ch1_line2d', dpi=350, transparent=False)
```
