## Annotation

```python
from matplotlib.text import Annotation
```

```python
annot = Annotation('Annotation Text 1', xy=(0.2,0.4), xytext=(0.5,0.8), arrowprops=dict(width=1))

fig2, ax2 = plt.subplots()

ax2._add_text(annot)
```

```python
ax2.texts[0].set_bbox(dict(facecolor='yellow', boxstyle='circle', ls='dashed', pad=0.4, alpha=0.7))
fig2
```

```python
ax2.texts[0].arrow_patch
```

```python
ax2.texts[0].arrow_patch.set_linewidth(2)
ax2.texts[0].arrow_patch.set_arrowstyle('-[')

fig2
```

# Legends

```python
from matplotlib.legend import Legend
from matplotlib.patches import Rectangle
```

```python
rec1 = Rectangle((0.2,0), 0.1, 0.3, label='L1', color='red')
rec2 = Rectangle((0.4,0), 0.1, 0.5, label='L2', color='green')
rec3 = Rectangle((0.6,0), 0.1, 0.7, label='L9', color='blue')

fig3, ax3 = plt.subplots()

ax3.add_patch(rec1)
ax3.add_patch(rec2)
ax3.add_patch(rec3)

ax3.patches[0]
```

### Method 1 to add legend

```python
legend = Legend(ax3, handles=[ax3.patches[0],ax3.patches[1],ax3.patches[2]],
               labels=[ax3.patches[0].get_label(), ax3.patches[1].get_label(), ax3.patches[2].get_label(),])

ax3.add_artist(legend)
fig3
```

### Method 2 to add legend

```python
handles, labels = ax3.get_legend_handles_labels()
legend = Legend(ax3, handles=handles, labels=labels)

ax3.add_artist(legend)
fig3
```

### Method 3 to add legend

```python
handles, labels = ax.get_legend_handles_labels()

ax3.legend(handles=handles, labels=labels)

fig3
```
