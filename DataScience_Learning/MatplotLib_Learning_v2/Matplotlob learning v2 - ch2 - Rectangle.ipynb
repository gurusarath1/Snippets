```python
import numpy as np
from matplotlib.patches import Rectangle
from matplotlib.patches import FancyBboxPatch
import matplotlib.pyplot as plt

%matplotlib inline
```

# Rectangle

```python
rec1 = Rectangle((1,3), 0.2, 0.8)
rec2 = Rectangle((2,0), 0.9, 0.7)

fig, ax = plt.subplots()
ax.add_patch(rec1)
ax.add_patch(rec2)
```

```python
ax.autoscale()
fig
```

```python
ax.patches
```

```python
rec1 = ax.patches[0]
rec2 = ax.patches[1]
```

```python
rec1.set_alpha(0.2)
fig
```

```python
for r in ax.patches:
    r.set_edgecolor('red')
    r.set_ls('dashed')
    r.set_lw(5)
    r.set_hatch('\\')
    
fig
```

```python
rec2.set_sketch_params(0.9)
fig
```

```python
rec2.set_clip_on(False)
ax.set_xlim(0, 2.5)
fig
```
