```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Part 1 - Visual Styles

```python
fig, ax = plt.subplots()
```

```python
ax.get_children()
```

```python
ax.spines
```

```python
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


ax.spines['bottom'].set_color('red')


fig
```

```python
ax.set_xlabel('X axis label')

fig
```

```python
x_label = ax.xaxis.get_label() # This returns a text artist

x_label.set_size(20)
x_label.set_fontweight('bold')
x_label.set_bbox(dict(facecolor='green', boxstyle='rarrow', edgecolor='red', linestyle='dashed', linewidth=2))
x_label.set_x(0.4)

fig
```

```python
fig2, ax2 = plt.subplots()

x = ax2.xaxis
y = ax2.yaxis

```

```python
for t in x.get_major_ticks():
    t.label1On = True
    t.label2On = True
    t.tick1On = True
    t.tick2On = True
    t.label2.set_visible(True)
    t.tick2line.set_visible(True)
    
    t.label2.set_color('red')
    t.label1.set_color('red')
    
    t.tick2line.set_marker('*')
    t.tick2line.set_markersize(15)

fig2
```

### Method 2 to format ticks

```python
fig3, ax3 = plt.subplots()

ax3.tick_params(axis='x',
               labelcolor='red',
               labelsize=14,
               top='On',
               labeltop='On',
               bottom='Off',
               labelbottom='Off')
```

## Part 2 - Range, Locating and formatting

### Setting xy ranges

```python
fig4, ax4 = plt.subplots()
```

```python
#Method1
ax4.axis([0,10, 0,15])

fig4
```

```python
#Method2
ax4.set_xlim(0,5)
ax4.set_ylim(0,9)


fig4
```

### Setting tick positions

### Locators

```python
from matplotlib.ticker import AutoLocator, MaxNLocator, LinearLocator, MultipleLocator, FixedLocator, NullLocator
```

```python
#Auto Locator
auto_loc = AutoLocator() # This is the default locator

ax4.xaxis.set_major_locator(auto_loc)

fig4
```

```python
# Max N locator
max_loc = MaxNLocator(10) # Maxinum number of intervals we want to see

ax4.xaxis.set_major_locator(max_loc)

fig4
```

```python
#Liner locator
lin_loc = LinearLocator(7)

ax4.xaxis.set_major_locator(lin_loc) # Sets the number of ticks

fig4
```

```python
# Multiple locator

ax4.set_xlim(0,20)

mul_loc = MultipleLocator(3) #Show every 3 ticks

ax4.xaxis.set_major_locator(mul_loc)

fig4
```

```python
# Fixed locator - Method 1

fix_loc = FixedLocator(locs=[0,1,5,11,12,18])

ax4.xaxis.set_major_locator(fix_loc)

fig4
```

```python
# Fixed locator - Method 2

ax4.set_xticks([0,1,2,10,18])

fig4
```

```python
# Null locator
ax4.xaxis.set_major_locator(NullLocator())

fig4
```

### Tick formatters # Formats the text

```python
ax4.xaxis.set_major_locator(auto_loc)
ax4.set_xlim(0,5)
```

```python
from matplotlib.ticker import NullFormatter, FixedFormatter, StrMethodFormatter, ScalarFormatter, PercentFormatter
```

```python
fig4
```

```python
null_form = NullFormatter()

ax4.xaxis.set_major_formatter(null_form)
fig4
```

```python
str_form = StrMethodFormatter('A{x:.2f}C')

ax4.xaxis.set_major_formatter(str_form)
fig4
```

```python
per_form = PercentFormatter(xmax = 10)

ax4.xaxis.set_major_formatter(per_form)
fig4
```
